```python
def sep_first(df):
    # new data frame with split value columns
    data_names = df["name"].str.split(" ", expand = True).copy()
    frames = [df, data_names]
    df = pd.concat(frames, axis = 1)
    df = column_rename(df,0,"first")
    df = column_rename(df,1,"second")
    df = column_rename(df,2,"third")
    df = column_rename(df,3,"fourth")
    return df

def column_rename(df,old,new):
    df = df.rename(columns = {old: new})
    return df
```
