```python
import csv
import pandas as pd
from pandas import DataFrame
import statsmodels.api as sm
import numpy as np
import warnings

warnings.filterwarnings('ignore')
```

```python
# Regression Results with Donation Change AS ABSOLUTE NUMBERS
a = "~/Documents/GitHub/CampaignDonations/Disney_Donor_Vote.csv"
df = pd.read_csv(a)
```

```python
df = df.drop(columns = ["person", "name_x" ,"party_x", "party_y", "Last_name" , "Suffix" , "LastName", "ID_NameParty", "Pure_name" , "FirstName" , "first" , "last", "religion" ,"death" , "birthplace", "deathplace"])
df.head()
```

```python
# Defining Function for Data preparation and OLS Regression Results

# Returns OLS Results 
def OLS_Results_function(independent, dependent): 
    X = independent
    Y = dependent
    X = sm.add_constant(X) # adding a constant
    model = sm.OLS(Y, X).fit()
    predictions = model.predict(X) 
    print_model = model.summary()
    print(print_model)
```

```python
# First Regression: Impact of VoteAlong113thRatio on change in donations between 113th and 114th Session
   
VoteAlong = "VoteAlong113thRatio"
PartyShort = "party_short"
DonationsA = "donations2014"
DonationsB = "donations2012"
SessionA =  "Session_113" 
SessionB = "Session_114"

        
df_First = df[[ "VoteAlong113thRatio","party_short","donations2012","donations2014","Session_113" ,"Session_114"]]
df_First = df_First.dropna(subset=[SessionA])
df_First = df_First.dropna(subset=[SessionB])
df_First= df_First.fillna(0)
df_First["Donation_Change_Absolute"] = 0

for x in range(len(df_First)):
    df_First["Donation_Change_Absolute"].iloc[x] = df_First[DonationsA].iloc[x] - df_First[DonationsB].iloc[x] 

df_First_D = df_First[df_First.party_short == "D"]
df_First_R = df_First[df_First.party_short == "R"]

```

```python
# Democrats VoteAlong113th

independent = df_First_D[["VoteAlong113thRatio"]]
dependent = df_First_D[["Donation_Change_Absolute"]]

OLS_Results_function(independent, dependent)
```

```python
# Republicans VoteAlong113th
independent = df_First_R[["VoteAlong113thRatio"]]
dependent = df_First_R[["Donation_Change_Absolute"]]

OLS_Results_function(independent, dependent)
```
