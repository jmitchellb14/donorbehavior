```python
df.drop(df[df.Number_Sessions_Served != 4.0].index, inplace=True)
recode = ['donations2012', 'donations2014', 'donations2016','donations2018']
df[recode] = df[recode].fillna(0)
```

```python
df_forlong = df[['person',
                "l2012", "l2014",'l2016',"l2018",
                "d2012", "d2014",'d2016',"d2018",
                "party_short"]].copy()
```

```python
df_forlong["person"] = df.index
df_long = pd.wide_to_long(df_forlong, ["d", "l"], i="person", j="Year")

df_long.rename(columns = {'d':"donations_dollars", "l":"partyline_percent"}, inplace = True)
```
