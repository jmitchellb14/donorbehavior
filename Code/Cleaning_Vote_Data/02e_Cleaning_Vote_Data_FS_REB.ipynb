```python
# Definition of Functions 

def my_function(link, bill): 
  a = "~/Documents/GitHub/CampaignDonations/External_Input_Data/Voting_Data/Vote_Results_FinancialSector_Real_Estate_Business_GovTrack/" + link
  df = pd.read_csv(a)
  df_Democratic = df[df["party"] == "Democrat"]
  df_Republican = df[df["party"] == "Republican"]
  df_Democratic["D_MajorityVote"] = df_Democratic["vote"].max()
  df_Republican["R_MajorityVote"] = df_Republican["vote"].max()
  # Getting Dummy for voting along party-line
  # As our main independet variable is a percetage of votes along party line, all independent candidates are dropped 
  df_Democratic["VoteAlongPartyLine"] = 0
  df_Republican["VoteAlongPartyLine"] = 0
  for x in range(len(df_Democratic)):
    if df_Democratic["vote"].iloc[x] == df_Democratic["D_MajorityVote"].iloc[x]:
     df_Democratic["VoteAlongPartyLine"].iloc[x] = 1
  for x in range(len(df_Republican)):
    if df_Republican["vote"].iloc[x] == df_Republican["R_MajorityVote"].iloc[x]: 
     df_Republican["VoteAlongPartyLine"].iloc[x] = 1

  # Merge both party-specific datasets into one 
  # First: drop party-specific columns 
  df_Democratic = df_Democratic.drop(columns = ["D_MajorityVote"])
  df_Republican = df_Republican.drop(columns = ["R_MajorityVote"])
  # Bind them back together
  df = df_Democratic.append(df_Republican)
  b = "VoteAlong" + bill
  df.columns = df.columns.str.replace('VoteAlongPartyLine', b)
  return(df)

# This Function returns a list with all unique MoH that votes at least once in the timeframe observed
def Unique_MoHs_function(MoH): 
    MoH = MoH.drop_duplicates()
    MoH = pd.DataFrame(MoH)
    MoH.columns = ['person', 'name']
    Unique_MoHs = MoH
    return(Unique_MoHs)
    
```

```python
# 113th Congress 

df1 = my_function("113th/congress_votes_113-2013_h487.csv", "1")
a1 = df1[["person", "name"]] 
b1 = df1.drop(columns = ["state", "district", "vote",  "name", "party"])

df2 = my_function("113th/congress_votes_113-2014_h98.csv", "2")
a2 = df2[["person", "name"]] 
b2 = df2.drop(columns = ["state", "district", "vote",  "name", "party"])

df3 = my_function("113th/congress_votes_113-2014_h404.csv", "3")
a3 = df3[["person", "name"]] 
b3 = df3.drop(columns = ["state", "district", "vote",  "name", "party"])




List_MoHs = a1.append(a2.append(a3))
Unique_MoHs = Unique_MoHs_function(List_MoHs)


All_MoHs_Merge = pd.merge(Unique_MoHs , b1, how='left')
All_MoHs_Merge = pd.merge(All_MoHs_Merge , b2, how='left')
All_MoHs_Merge = pd.merge(All_MoHs_Merge , b3, how='left')



All_MoHs_Merge["VoteAlong113thRatio_FS_REB"] = All_MoHs_Merge["VoteAlong1"]  + All_MoHs_Merge["VoteAlong2"] + All_MoHs_Merge["VoteAlong3"] 

All_MoHs_Merge["VoteAlong113thRatio_FS_REB"] = All_MoHs_Merge["VoteAlong113thRatio_FS_REB"]/3

All_MoHs_Merge_113th_FS_REB = All_MoHs_Merge


list = ["VoteAlong1", "VoteAlong2", "VoteAlong3" ]
All_MoHs_Merge_113th_FS_REB =  All_MoHs_Merge_113th_FS_REB.drop(columns = list)

All_MoHs_Merge_113th_FS_REB.head()
```
