```python
import csv
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Definition of Functions 

def my_function(link, bill): 
  a = "~/Documents/GitHub/CampaignDonations/Data/Voting_Data/Vote_Results_Television&Film_GovTrack/" + link
  df = pd.read_csv(a)
  df_Democratic = df[df["party"] == "Democrat"]
  df_Republican = df[df["party"] == "Republican"]
  df_Democratic["D_MajorityVote"] = df_Democratic["vote"].max()
  df_Republican["R_MajorityVote"] = df_Republican["vote"].max()
  # Getting Dummy for voting along party-line
  # As our main independet variable is a percetage of votes along party line, all independent candidates are dropped 
  df_Democratic["VoteAlongPartyLine"] = 0
  df_Republican["VoteAlongPartyLine"] = 0
  for x in range(len(df_Democratic)):
    if df_Democratic["vote"].iloc[x] == df_Democratic["D_MajorityVote"].iloc[x]:
     df_Democratic["VoteAlongPartyLine"].iloc[x] = 1
  for x in range(len(df_Republican)):
    if df_Republican["vote"].iloc[x] == df_Republican["R_MajorityVote"].iloc[x]: 
     df_Republican["VoteAlongPartyLine"].iloc[x] = 1

  # Merge both party-specific datasets into one 
  # First: drop party-specific columns 
  df_Democratic = df_Democratic.drop(columns = ["D_MajorityVote"])
  df_Republican = df_Republican.drop(columns = ["R_MajorityVote"])
  # Bind them back together
  df = df_Democratic.append(df_Republican)
  b = "VoteAlong" + bill
  df.columns = df.columns.str.replace('VoteAlongPartyLine', b)
  return(df)

# This Function returns a list with all unique MoH that votes at least once in the timeframe observed
def Unique_MoHs_function(MoH): 
    MoH = MoH.drop_duplicates()
    MoH = pd.DataFrame(MoH)
    MoH.columns = ['person', 'name']
    Unique_MoHs = MoH
    return(Unique_MoHs)
    
```

```python
# 113th Congress 

df1 = my_function("113th/congress_votes_113-2014_h41.csv", "2014_h41")
a1 = df1[["person", "name"]] 
b1 = df1.drop(columns = ["state", "district", "vote",  "name", "party"])


df2 = my_function("113th/congress_votes_113-2014_h269.csv", "2014_h269")
a2 = df2[["person", "name"]] 
b2 = df2.drop(columns = ["state", "district", "vote",  "name", "party"])


df3 = my_function("113th/congress_votes_113-2014_h544.csv", "2014_h544")
a3 = df3[["person", "name"]] 
b3 = df3.drop(columns = ["state", "district", "vote",  "name", "party"])


df4 = my_function("113th/congress_votes_113-2014_h563.csv", "2014_h563")
a4 = df4[["person", "name"]] 
b4 = df4.drop(columns = ["state", "district", "vote",  "name"])


List_MoHs = a1.append(a2.append(a3.append(a4)))
Unique_MoHs = Unique_MoHs_function(List_MoHs)


All_MoHs_Merge = pd.merge(Unique_MoHs , b1, how='left')
All_MoHs_Merge = pd.merge(All_MoHs_Merge , b2, how='left')
All_MoHs_Merge = pd.merge(All_MoHs_Merge , b3, how='left')
All_MoHs_Merge = pd.merge(All_MoHs_Merge , b4, how='left')


All_MoHs_Merge["VoteAlong113thRatio"] = All_MoHs_Merge["VoteAlong2014_h41"] + All_MoHs_Merge["VoteAlong2014_h269"] + All_MoHs_Merge["VoteAlong2014_h544"] +  All_MoHs_Merge["VoteAlong2014_h563"]
All_MoHs_Merge["VoteAlong113thRatio"] = All_MoHs_Merge["VoteAlong113thRatio"]/4

All_MoHs_Merge_113th = All_MoHs_Merge


list = ["VoteAlong2014_h41","VoteAlong2014_h269", "VoteAlong2014_h544", "VoteAlong2014_h563"]
All_MoHs_Merge_113th =  All_MoHs_Merge_113th.drop(columns = list)

All_MoHs_Merge_113th.head()
```

```python
# 114th Session


df1 = my_function("114th/congress_votes_114-2015_h227.csv", "2015_h277")
a1 = df1[["person", "name"]] 
b1 = df1.drop(columns = ["state", "district", "vote",  "name", "party"])


df2 = my_function("114th/congress_votes_114-2016_h101.csv", "2016_h101")
a2 = df2[["person", "name"]] 
b2 = df2.drop(columns = ["state", "district", "vote",  "name", "party"])


df3 = my_function("114th/congress_votes_114-2016_h294.csv", "2016_h294")
a3 = df3[["person", "name"]] 
b3 = df3.drop(columns = ["state", "district", "vote",  "name"])


List_MoHs = a1.append(a2.append(a3))
Unique_MoHs = Unique_MoHs_function(List_MoHs)


All_MoHs_Merge = pd.merge(Unique_MoHs , b1, how='left')
All_MoHs_Merge = pd.merge(All_MoHs_Merge , b2, how='left')
All_MoHs_Merge = pd.merge(All_MoHs_Merge , b3, how='left')


All_MoHs_Merge["VoteAlong114thRatio"] = All_MoHs_Merge["VoteAlong2015_h277"] + All_MoHs_Merge["VoteAlong2016_h101"] + All_MoHs_Merge["VoteAlong2016_h294"] 
All_MoHs_Merge["VoteAlong114thRatio"] = All_MoHs_Merge["VoteAlong114thRatio"]/3

All_MoHs_Merge_114th = All_MoHs_Merge

list = ["VoteAlong2015_h277", "VoteAlong2016_h101", "VoteAlong2016_h294"]
All_MoHs_Merge_114th =  All_MoHs_Merge_114th.drop(columns = list)


All_MoHs_Merge_114th.head()
```

```python
# 115th Session


df1 = my_function("115th/congress_votes_115-2017_h210.csv", "2017_h210")
a1 = df1[["person", "name"]] 
b1 = df1.drop(columns = ["state", "district", "vote",  "name", "party"])

df2 = my_function("115th/congress_votes_115-2017_h378.csv", "2017_h378")
a2 = df2[["person", "name"]] 
b2 = df2.drop(columns = ["state", "district", "vote",  "name", "party"])

df3 = my_function("115th/congress_votes_115-2017_h413.csv", "2017_h413")
a3 = df3[["person", "name"]] 
b3 = df3.drop(columns = ["state", "district", "vote",  "name", "party"])

df4 = my_function("115th/congress_votes_115-2018_h127.csv", "2018_h127")
a4 = df4[["person", "name"]] 
b4 = df4.drop(columns = ["state", "district", "vote",  "name", "party"])

df5 = my_function("115th/congress_votes_115-2018_h434.csv", "2018_h434")
a5 = df5[["person", "name"]] 
b5 = df5.drop(columns = ["state", "district", "vote",  "name"])


List_MoHs = a1.append(a2.append(a3.append(a4.append(a5))))
Unique_MoHs = Unique_MoHs_function(List_MoHs)


All_MoHs_Merge = pd.merge(Unique_MoHs , b1, how='left')
All_MoHs_Merge = pd.merge(All_MoHs_Merge , b2, how='left')
All_MoHs_Merge = pd.merge(All_MoHs_Merge , b3, how='left')
All_MoHs_Merge = pd.merge(All_MoHs_Merge , b4, how='left')
All_MoHs_Merge = pd.merge(All_MoHs_Merge , b5, how='left')


All_MoHs_Merge["VoteAlong115thRatio"] = All_MoHs_Merge["VoteAlong2017_h210"] + All_MoHs_Merge["VoteAlong2017_h378"] + All_MoHs_Merge["VoteAlong2017_h413"]  + All_MoHs_Merge["VoteAlong2018_h127"] + All_MoHs_Merge["VoteAlong2018_h434"]  
All_MoHs_Merge["VoteAlong115thRatio"] = All_MoHs_Merge["VoteAlong115thRatio"]/5

All_MoHs_Merge_115th = All_MoHs_Merge


list = ["VoteAlong2017_h210","VoteAlong2017_h378","VoteAlong2017_h413","VoteAlong2018_h127","VoteAlong2018_h434"]
All_MoHs_Merge_115th =  All_MoHs_Merge_115th.drop(columns = list)


All_MoHs_Merge_115th.head()
```

```python
# 116th Session


df1 = my_function("116th/congress_votes_116-2019_h118.csv", "2019_h118")
a1 = df1[["person", "name"]] 
b1 = df1.drop(columns = ["state", "district", "vote",  "name", "party"])


df2 = my_function("116th/congress_votes_116-2020_h109.csv", "2020_h109")
a2 = df2[["person", "name"]] 
b2 = df2.drop(columns = ["state", "district", "vote",  "name"])



List_MoHs = a1.append(a2)
Unique_MoHs = Unique_MoHs_function(List_MoHs)


All_MoHs_Merge = pd.merge(Unique_MoHs , b1, how='left')
All_MoHs_Merge = pd.merge(All_MoHs_Merge , b2, how='left')



All_MoHs_Merge["VoteAlong116thRatio"] = All_MoHs_Merge["VoteAlong2019_h118"] + All_MoHs_Merge["VoteAlong2020_h109"] 
All_MoHs_Merge["VoteAlong116thRatio"] = All_MoHs_Merge["VoteAlong116thRatio"]/2

All_MoHs_Merge_116th = All_MoHs_Merge

list = ["VoteAlong2019_h118", "VoteAlong2020_h109"]
All_MoHs_Merge_116th =  All_MoHs_Merge_116th.drop(columns = list)


All_MoHs_Merge_116th.head()
```

```python
# Merge different sessions into one data set 

c1 = All_MoHs_Merge_113th[["person", "name", "party"]] 
c2 = All_MoHs_Merge_114th[["person", "name", "party"]] 
c3 = All_MoHs_Merge_115th[["person", "name", "party"]] 
c4 = All_MoHs_Merge_116th[["person", "name", "party"]] 

    
List_MoHs_All_Sessions = c1.append(c2.append(c3.append(c4)))

List_MoHs_All_Sessions = List_MoHs_All_Sessions.drop_duplicates()
List_MoHs_All_Sessions = pd.DataFrame(List_MoHs_All_Sessions)
List_MoHs_All_Sessions.columns = ['person', 'name', 'party']
Unique_MoHs_All_Sessions = List_MoHs_All_Sessions


MoHs_VoteRatios_113_116 = pd.merge(Unique_MoHs_All_Sessions , All_MoHs_Merge_113th, how='left')
MoHs_VoteRatios_113_116 = pd.merge(MoHs_VoteRatios_113_116 , All_MoHs_Merge_114th, how='left')
MoHs_VoteRatios_113_116 = pd.merge(MoHs_VoteRatios_113_116 , All_MoHs_Merge_115th, how='left')
MoHs_VoteRatios_113_116 = pd.merge(MoHs_VoteRatios_113_116 , All_MoHs_Merge_116th, how='left')

MoHs_VoteRatios_113_116.head()
```

```python
# Export Dataframe into CSV

compression_opts = dict(method='zip',
                        archive_name='MoHs_VoteRatios_113_116_Film.csv')  
MoHs_VoteRatios_113_116.to_csv('MoHs_VoteRatios_113_116_Film.zip', index=False,
          compression=compression_opts)  
```
